# 作業2：

## 想像你經營一個自由載客車隊，你希望能透過數據分析以提升業績，請你思考並描述你如何規劃整體的分析/解決方案：


### 1. 核心問題為何 (tips：如何定義 「提升業績 & 你的假設」)
    提升業績：
        1. 載客次數增加
        2. 減少路上空車率
        3. 提升載客品質(顧客叫車流程改善、減少顧客等車時間、顧客評價分數上升)
    假設：
        1. 某些日子或是時段客戶搭車的需求較多(如：假日傍晚)？
        2. 某些地點載客機會比較高
        3. 天氣是否影響載客人數
        4. 顧客評分上升會提升業績
        
### 2. 資料從何而來 (tips：哪些資料可能會對你想問的問題產生影響 & 資料如何蒐集)
        司機駕駛紀錄軌跡(GPS定位追蹤)、中央氣象局api串接天氣資訊、叫車app顧客所留評分評語

### 3. 蒐集而來的資料型態為何
        結構化資料：行車、接送時間、天氣資訊、顧客評分評語
        非結構化資料：GPS軌跡定位資料

### 4. 你要回答的問題，其如何評估 (tips：你的假設如何驗證)
        描述性統計：將日期和星期與載客數做長條圖、將天氣和載客數座長條圖
        地理座標圖：將所有載客車輛呈現在地圖上用顏色區分載客人數，顏色越深代表載客人數越多該區載客機會較高
        相關性分析：將顧客評分和業績做相關性統計分析
